```python
x=Bos.iloc[:,0:-1]
x.head()
```

```python
x.shape
```

```python
y=Bos.iloc[:,-1]
y.head()
```

```python
y.shape
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=999)
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
y_train.shape
```

```python
y_test.shape
```

```python
lm=LinearRegression()
```

```python
lm.fit(x_train,y_train)
```

```python
lm.coef_
```

```python
lm.intercept_
```

```python
lm.score(x_train,y_train)
```

```python
# Predict the values:-
pred=lm.predict(x_test)
print("Predicted result price: ",pred)
print("Actual Price: ",y_test)
```

```python
print("Error :")
print("Mean Absolute Error: ", mean_absolute_error(y_test,pred))
print("Mean Squared Error: ", mean_squared_error(y_test,pred))
print("Root Mean Squared Error: ",np.sqrt(mean_squared_error(y_test,pred)))
```

```python
# r2 score is Coefficient of determination
# It is the proportion of variance in the dependent variable that is predictible from the independent variable's
# change. ie: proportion of change in Y whenever X is being changed.
from sklearn.metrics import r2_score
print(r2_score(y_test,pred))
```
