# PCA (Principal Component Analysis):-

```python
# Due to the high number of columns ie: 60 columns we are not able to generate a very good accuracy score
# The PCA Analysis merges the columns which are higly correlated into 1 and take the mean of all columns which are merged
# This way the number of columns will be reduced without any loss in data and model can easily learn with less columns
```

```python
from sklearn.decomposition import PCA
```

```python
pca=PCA(n_components=10)
```

```python
y=df.iloc[:,-1]
```

```python
x=pca.fit_transform(x)
```

```python
x.shape
```

```python
pd.DataFrame(data=x)
```

```python
y
```

# Label Encoder:-

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le=LabelEncoder()    # Converting The Target Variable Y into 0 & 1 inplace of Rocks & Mines
y=le.fit_transform(y)
y
```

# Scaling (Standard Scaler):-

```python
from sklearn.preprocessing import StandardScaler
```

```python
scale=StandardScaler()
```

```python
x=scale.fit_transform(x)
```

# Evaluating Model using different Classifiers:- 

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.22,random_state=100)
```

```python
model=LogisticRegression()
```

```python
x_train.shape
```

```python
y_train.shape
```

```python
x_test.shape
```

```python
y_test.shape
```

```python
model.fit(x_train,y_train)
```

```python
logpred=model.predict(x_test)
```

```python
print(accuracy_score(logpred,y_test))
```

```python
print(confusion_matrix(logpred,y_test))
```

```python
print(classification_report(logpred,y_test))
```

### Decision Tree Classifier:-

```python
dtc=DecisionTreeClassifier()
dtc.fit(x_train,y_train)
```

```python
dtcpred=dtc.predict(x_test)
```

```python
print(accuracy_score(dtcpred,y_test))
```
